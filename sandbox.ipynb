```python
%matplotlib inline

import pandas as pd
import numpy as np
import unittest
import glob, os
import csv
import re
import argparse
import matplotlib.pyplot as plt


def plot_sandbox():
    fig = plt.figure(figsize=(20, 10))
    plt.plot(np.arange(10), np.random.randn(10), 'b--')
    plt.axhline(1)
    plt.axvline(5)
    plt.show()
    

# percentage change
def pandas_pctchange():
    s1 = pd.Series(np.random.randint(10, size=10))
    print(list(s1))
    print(s1.pct_change())
    print(s1.pct_change(2))
    

# covariance and correlation
def pandas_covcorr():
    s1 = pd.Series(np.random.randint(20, size=10))
    print(list(s1))
    s2 = pd.Series(np.random.randint(20, size=10))
    print(list(s2))
    print(s1.cov(s2))
    print(s1.corr(s2))
    print(s1.corr(s1))

    
def pandas_rank():
    df = pd.DataFrame(np.random.randn(8, 3), columns=list('abc'))
    print(df)
    print(df.rank())
    print(df.rank(ascending=False))
    print(df.rank(axis=1))

    
if __name__ == "__main__":
    pandas_rank()
```

```python
%matplotlib inline

import pandas as pd
import numpy as np

s = pd.Series(np.random.randn(100), index=pd.date_range('2010-1-1', periods=100))
s = s.cumsum()
r = s.rolling(window=10)
s.plot(style='k--', figsize=(30, 15), fontsize=15)
r.mean().plot(style='b')
```
